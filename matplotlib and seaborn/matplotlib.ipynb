## 제일 먼저 style부터 정한다
#### style을 나중에 지정하면 이전에 설정해둔 정보가 다 덮어씌워짐

```python
plt.style.use('ggplot')
plt.figure(figsize=(7,5)) #눈에 안 보이는 영역. 전체 크기는 여기서 정한다.
plt.rcParams['font.family']='NanumGothic'
plt.axes() #눈에 보이는 영역
plt.title('타이틀')
plt.ylabel('ylabel')
plt.xlabel('xlabel')
plt.grid(True)
#plt.yticks([])
#plt.xticks([])
plt.plot([[1,2,3],[4,5,6]])
```
