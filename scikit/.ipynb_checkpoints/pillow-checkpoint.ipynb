만약 이미지를 수만장씩 가져오려면? 폴더 하나를 통째로 읽어야 함.

```python
import os
path = './images'
```

```python
temp = []
for i in os.listdir(path):
    s = Image.open(os.path.join(path,i))
    t = np.array(s)
    temp .append(t)
```
